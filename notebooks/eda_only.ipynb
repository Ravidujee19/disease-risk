```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Show plots inline
%matplotlib inline
```

```python
# Load dataset
data = pd.read_csv("/content/drive/MyDrive/disease_risk.csv")
print("Shape:", data.shape)
data.head()
```

```python
# Overview of dataset
data.info()

# Summary statistics
data.describe(include="all").transpose()
```

```python
# Remove the survey code
if data.columns[0].lower() == "survey_code":
    data = data.drop(columns=[data.columns[0]])

print("Shape after dropping first column:", data.shape)
data.head()
```

```python
# Missing Values Heatmap
plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Target Variable Distribution
if "target" in data.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x="target", data=data, palette="coolwarm")
    plt.title("Target Variable Distribution")
    plt.show()
```

```python
# Numeric Features Distribution
num_cols = data.select_dtypes(include="number").columns.tolist()

data[num_cols].hist(figsize=(15,15), bins=30, edgecolor="black")
plt.title("Numeric Feature Distributions", fontsize=16)
plt.show()
```

```python
# Correlation Heatmap for Numaric data type
plt.figure(figsize=(12,8))
sns.heatmap(data[num_cols].corr(), annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
# Categorical Features Countplots
cat_cols = data.select_dtypes(include="object").columns.tolist()

for col in cat_cols[:6]:  # Show first 6
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=data, hue=col, palette="muted", legend=False)
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Boxplots (Outlier Visualization) ---
for col in num_cols[:6]:  # Show first 6
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[col], color="skyblue")
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Pairplot for Selected Features
sampled = data.sample(n=2000, random_state=42) if len(data) > 2000 else data

sns.pairplot(sampled[num_cols[:5] + ['target']],
             hue="target", diag_kind="kde", palette="husl")
plt.suptitle("Pairplot of Selected Features vs Target", y=1.02, fontsize=16)
plt.show()

```

```python
# Target vs Numeric Features (Boxplots)
if "target" in data.columns:
    for col in num_cols[:8]:
        plt.figure(figsize=(6,4))
        sns.boxplot(x="target", y=col, data=data, hue="target", palette="Set2", legend=False)
        plt.title(f"{col} distribution across Target classes")
        plt.show()
```

```python
# Target vs Categorical Features (Countplots)
if "target" in data.columns:
    for col in cat_cols[:8]:
        plt.figure(figsize=(6,4))
        sns.countplot(x=col, data=data, hue="target", palette="coolwarm")
        plt.title(f"{col} vs Target")
        plt.show()
```

```python
if "target" in data.columns:
    num_cols = data.select_dtypes(include="number").columns.tolist()
    num_cols = [c for c in num_cols if c != "target"]

    # Numeric version of target
    tgt = pd.to_numeric(data["target"], errors="coerce")

    if tgt.isna().all():
        codes, uniques = pd.factorize(data["target"], sort=True)
        tgt = pd.Series(codes, index=data.index, name="target_num")
        print("Encoded target labels → codes:")
        for i, u in enumerate(uniques):
            print(f"  {u!r} -> {i}")
    else:
        tgt = tgt.fillna(tgt.mode().iloc[0])
        tgt.name = "target_num"

    # Remove constant columns
    num_cols = [c for c in num_cols if data[c].nunique() > 1]

    # Compute Pearson correlation
    corr_with_target = data[num_cols].apply(lambda s: s.corr(tgt))
    corr_with_target = corr_with_target.dropna().sort_values(ascending=False)

    # Plot
    plt.figure(figsize=(8, max(4, 0.35 * len(corr_with_target))))
    sns.barplot(x=corr_with_target.values, y=corr_with_target.index)
    plt.title("Correlation of Numeric Features with Target (encoded)")
    plt.xlabel("Pearson r")
    plt.ylabel("")
    plt.show()
else:
    print("Column 'target' not found in the dataframe.")
```

```python
# KDE Plots for Distribution by Target
if "target" in data.columns:
    for col in num_cols[:6]:
        plt.figure(figsize=(6,4))
        sns.kdeplot(data=data, x=col, hue="target", fill=True, common_norm=False, alpha=0.5)
        plt.title(f"Density plot of {col} by Target")
        plt.show()
```
